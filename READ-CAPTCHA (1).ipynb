```python
import cv2

def plot_(img1, img2):
    # Perform adaptive thresholding on the input images
    thresh_img1 = cv2.adaptiveThreshold(img1, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 145, 0)
    thresh_img2 = cv2.adaptiveThreshold(img2, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 145, 0)

    # Perform morphological closing on the thresholded images
    kernel_close = np.ones((5, 2), np.uint8)
    close_img1 = cv2.morphologyEx(thresh_img1, cv2.MORPH_CLOSE, kernel_close)
    close_img2 = cv2.morphologyEx(thresh_img2, cv2.MORPH_CLOSE, kernel_close)

    # Perform dilation on the closed images
    kernel_dilate = np.ones((2, 2), np.uint8)
    dilate_img1 = cv2.dilate(close_img1, kernel_dilate, iterations=1)
    dilate_img2 = cv2.dilate(close_img2, kernel_dilate, iterations=1)

    # Perform Gaussian blur on the dilated images
    gauss_img1 = cv2.GaussianBlur(dilate_img1, (1, 1), 0)
    gauss_img2 = cv2.GaussianBlur(dilate_img2, (1, 1), 0)

    # Find contours in the dilated images
    contours1, _ = cv2.findContours(dilate_img1, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    contours2, _ = cv2.findContours(dilate_img2, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Draw rectangles around each character in the images
    for contour in contours1:
        x, y, w, h = cv2.boundingRect(contour)
        if w > 5 and h > 5:  # Filter out small artifacts
            cv2.rectangle(img1, (x, y), (x + w, y + h), 255, 1)  # Draw rectangle on the original image

    for contour in contours2:
        x, y, w, h = cv2.boundingRect(contour)
        if w > 5 and h > 5:  # Filter out small artifacts
            cv2.rectangle(img2, (x, y), (x + w, y + h), 255, 1)  # Draw rectangle on the original image

    # Plot the images with rectangles around each character
    plt.figure(figsize=(15, 15))

    plt.subplot(3, 3, 1)
    plt.imshow(img1, cmap='gray')
    plt.title("Image 1 with Rectangles")
    plt.axis('off')

    plt.subplot(3, 3, 2)
    plt.imshow(thresh_img1, cmap='gray')
    plt.title("Thresholded Image 1")
    plt.axis('off')

    plt.subplot(3, 3, 3)
    plt.imshow(dilate_img1, cmap='gray')
    plt.title("Dilated Image 1")
    plt.axis('off')

    plt.subplot(3, 3, 4)
    plt.imshow(img2, cmap='gray')
    plt.title("Image 2 with Rectangles")
    plt.axis('off')

    plt.subplot(3, 3, 5)
    plt.imshow(thresh_img2, cmap='gray')
    plt.title("Thresholded Image 2")
    plt.axis('off')

    plt.subplot(3, 3, 6)
    plt.imshow(dilate_img2, cmap='gray')
    plt.title("Dilated Image 2")
    plt.axis('off')

    plt.tight_layout()
    plt.show()


# Example usage:
img1 = cv2.imread('C:/Users/MYTHILI/Desktop/captcha-3data/train/C1/00bAQwhAZU.jpg', cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread('C:/Users/MYTHILI/Desktop/captcha-3data/train/C2/00D95.png', cv2.IMREAD_GRAYSCALE)
plot_(img1, img2)
```

```python
from keras.models import Sequential 
from keras.layers import Dense
from keras.layers import Conv2D
from keras.layers import Flatten
from keras.layers import MaxPooling2D
from keras.layers import BatchNormalization

from keras.layers import Dropout
from keras.layers import Input
```

```python
def conv_layer (filterx) :
    
    model = Sequential()
    
    model.add(Conv2D(filterx, (3,3), padding = 'same', activation = 'relu'))
    model.add(BatchNormalization())
    model.add(Dropout(0.2))
    model.add(MaxPooling2D(pool_size = (2,2), padding = 'same'))
    
    return model
```

```python
def dens_layer (hiddenx) :
    
    model = Sequential()
    
    model.add(Dense(hiddenx, activation = 'relu'))
    model.add(BatchNormalization())
    model.add(Dropout(0.2))
    
    return model
```

```python
def cnn (filter1, filter2, filter3, hidden1, hidden2) :
    
    model = Sequential()
    model.add(Input((40, 20, 1,)))
    
    model.add(conv_layer(filter1))
    model.add(conv_layer(filter2))
    model.add(conv_layer(filter3))
    
    model.add(Flatten())
    model.add(dens_layer(hidden1))
    model.add(dens_layer(hidden2))
    
    model.add(Dense(19, activation = 'softmax'))
    
    model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
    
    return model
```
